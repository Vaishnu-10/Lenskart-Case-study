```python
social_media_df = pd.DataFrame.from_dict(social_media_data, orient='index', columns=['Percentage'])
plt.figure(figsize=(10, 6))
sns.barplot(x=social_media_df.index, y='Percentage', data=social_media_df)
plt.title('Lenskart Traffic Distribution - Social Media Platforms')
plt.xlabel('Social Media Platform')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
website_traffic_df = pd.DataFrame.from_dict(website_traffic_data, orient='index', columns=['Percentage'])
website_traffic_df = website_traffic_df.sort_values(by=['Percentage'], ascending=False)  # Sort by highest percentage first

plt.figure(figsize=(10, 6))
sns.barplot(x=website_traffic_df.index, y='Percentage', data=website_traffic_df)
plt.title('Lenskart Website Traffic Sources')
plt.xlabel('Traffic Source')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()
```
